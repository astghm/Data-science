```python
img = np.array(Image.open("Penguins_2.jpg"))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

** Use matplotlib to visualize the `img **

** Check the range of numbers of the picture **
** Numbers either are from 0-255 or 0-1(normalized) **
** If the picture is in range 0-255, normalize the picture by dividing by 255 **

** Be careful to not normalize more than once **

```python
img[:,:,0]
```

```python
img = img/255
```

** Find the standard deviation and mean of each channels RGB **

```python
print("Red Channel, (std, mean)" )

(img[:,:,0].std(),img[:,:,0].mean())
```

```python
print("Green Channel, (std, mean)" )

(img[:,:,1].std(),img[:,:,1].mean())
```

```python
print("Blue Channel, (std, mean)" )

(img[:,:,2].std(),img[:,:,2].mean())
```
